```python
# Q1 :


import numpy as np
import matplotlib.pyplot as plt


T = 1.0        
N = 1000       
dt = T / N    
t = np.linspace(0, T, N+1) 


dW = np.sqrt(dt) * np.random.randn(N) 


W = np.zeros(N+1)      
W[1:] = np.cumsum(dW)  


plt.figure(figsize=(10, 5))
plt.plot(t, W, label='Wiener Process Path')
plt.title("Simulated Standard Brownian Motion (Wiener Process)")
plt.xlabel("Time t")
plt.ylabel("W(t)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Q2 :


import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)

# Parameters
S0 = 100         
T = 1.0         
N = 1000         
dt = T / N       
t = np.linspace(0, T, N+1)  

alpha = 0.1     
sigma = 0.2      
num_paths = 5    


def simulate_gbm(S0, alpha, sigma, T, N):
    dt = T / N
    t = np.linspace(0, T, N+1)
    W = np.random.randn(N) * np.sqrt(dt)  
    W = np.insert(np.cumsum(W), 0, 0)     
    S = S0 * np.exp(sigma * W + (alpha - 0.5 * sigma**2) * t)
    return t, S


plt.figure(figsize=(10, 6))

for i in range(num_paths):
    t, S = simulate_gbm(S0, alpha, sigma, T, N)
    plt.plot(t, S, label=f"Path {i+1}")

plt.title("5 Simulated Paths of Geometric Brownian Motion")
plt.xlabel("Time t")
plt.ylabel("Stock Price S(t)")
plt.grid(True)
plt.legend()
plt.show()
```
